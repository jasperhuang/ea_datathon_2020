```
merged_df = pd.merge(left=all_features_df, right=label_df, on='user_id')
```

```
merged_df.head(5)
```

```
merged_df2 = merged_df[merged_df['num_days_sub'] < 360]
```

```
merged_df2.head(5)
```

```
def binarize(val):
  if val < 50:
    return 0
  else:
    return 1
```

```
merged_df2['label'] = merged_df2['num_days_sub'].apply(binarize)
```

```
merged_df2.head(5)
```

```
# def sessions_per_day(row):
#   return int(row['num_sessions_played']) / int(row['num_days_played'])
```

```
merged_df2['avg_sessions_per_day'] = merged_df2['num_sessions_played'] / merged_df2['num_days_played']
```

```
merged_df2.head(5)
```

```
merged_df3=merged_df2.drop(columns=['num_sessions_played','num_days_played'])
```

```
merged_df4=merged_df3.sort_values(by='label',ascending=True)
```

```
merged_df4
```

```
game_titles = ['MASS EFFECT: ANDROMEDA',
 'PLANTS VS ZOMBIES GARDEN WARFARE',
 'PLANTS VS ZOMBIES GARDEN WARFARE 2',
 'SIMCITY (2013)',
 'DEAD SPACE 3',
 'BATTLEFIELD HARDLINE',
 'TITANFALL 2',
 'DRAGON AGE: INQUISITION',
 'MASS EFFECT 1',
 'BATTLEFIELD #3',
 'SIMS 3',
 'CRYSIS 3',
 'BATTLEFRONT',
 'NFS 2016',
 'MASS EFFECT 3',
 'FIFA 16',
 'FIFA 17',
 'BATTLEFIELD 1',
 'SIMS 4',
 'BATTLEFIELD 4']
temp_df = merged_df4.copy()
for title in game_titles:
  temp_df[title] = (temp_df[title] > 0)*1
```

```
# temp_df.head(20)
```

```
#### final feature list
features_df = temp_df.drop(columns=['avg_sessions_per_day','user_id','label','num_days_sub']).fillna(0)
features_df.tail(5)
```

```
feature_names = features_df.columns.to_list()
feature_names
```

```
X = np.array(features_df.as_matrix()).astype(float)
```

```
X[X==np.inf] = 0
```

```
np.where(np.isinf(X))
```

```
y=np.array(merged_df4['label'])
```

```
y
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit(X).transform(X)
```

```
X_scaled
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
from sklearn import svm
cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
clf1 = LogisticRegression(C=1e5)
clf2 = svm.SVC(kernel='linear', C=1)
```

```
X
```

```
cross_val_score(clf1, X, y, cv=cv, scoring='f1')
```

```
cross_val_score(clf1, X_scaled, y, cv=cv, scoring='f1')
```
